```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
selected_feature = ['Genre', 'Description']

print(selected_feature)
```

```python
for feature in selected_feature:
    df[feature] = df[feature].fillna('')
```

```python
combined_feature = df['Genre']
```

```python
print(combined_feature)
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vector)
```

```python
similarity = cosine_similarity(feature_vector)
```

```python
feature_vector
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
Game_Name = input("Enter The Game Name:")
```

```python
list_of_all_titles = df['Game Name'].tolist()
print(list_of_all_titles)
```

```python
find_close_match = difflib.get_close_matches(Game_Name,list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
df.rename(columns={'Game Name': 'Name'}, inplace=True)
```

```python
df.head()
```

```python
index_of_the_game = df[df.Name == close_match]['sr.no'].values[0]
print(index_of_the_game)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_game]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_game = sorted (similarity_score, key=lambda x:x[1],reverse = True)
print(sorted_similar_game)
```

```python
print('Games Suggested For You: \n')

i = 1
for game in sorted_similar_game:
    index = game[0]
    title_from_index = df[df.index==index]['Name'].values[0]
    if(i<11):
        print(i,'.',title_from_index)
        i+=1
```

```python
Game_Name = input("Enter The Game Name:")
list_of_all_titles = df['Name'].tolist()
find_close_match = difflib.get_close_matches(Game_Name,list_of_all_titles)
close_match = find_close_match[0]
index_of_the_game = df[df.Name == close_match]['sr.no'].values[0]
similarity_score = list(enumerate(similarity[index_of_the_game]))
sorted_similar_game = sorted (similarity_score, key=lambda x:x[1],reverse = True)
print('Games Suggested For You: \n')

i = 1
for game in sorted_similar_game:
    index = game[0]
    title_from_index = df[df.index==index]['Name'].values[0]
    if(i<11):
        print(i,'.',title_from_index)
        i+=1
```
